```python
import pandas as pd
data = pd.read_csv("spotify_2023.csv", encoding="latin-1")
```

```python
#1. Завдання полягає в тому, щоб побудувати горизонтальну стовпчикову діаграму, на якій відображаються 10 найпопулярніших пісень на Spotify у 2023 році за кількістю прослуховувань.
import matplotlib.pyplot as plt
sort = data.sort_values(by="in_spotify_playlists",  ascending=False).head(10)
plt.figure(figsize=(10,4))
plt.barh(sort["track_name"], sort["in_spotify_playlists"], color = "purple")
plt.xlabel("Кількість слухачів")
plt.ylabel("Назва треку")
plt.title("10 найпопулярніших пісень на Spotify у 2023")
plt.gca().invert_yaxis()
plt.show()
```

```python
# 2. Розподіл жанрів найпопулярніших пісень: Створіть кругову діаграму або секторну діаграму, яка відображає розподіл жанрів серед найпопулярніших пісень у 2023 році. 
year = data["released_year"].value_counts()
top_year = year.head(6)
plt.figure(figsize=(8, 8))
plt.pie(top_year, labels=top_year.index, autopct='%1.1f%%', startangle=140)
plt.title('Роки виходу найбільш популярних пісень')
plt.show()
#Так як у нас немає стовпця з інформацією про жанр треків, я вирішила зробити діаграму по роках виходу набільш популярних пісень) 
```

```python
# 4. Рейтинг артистів на Spotify: Створіть стовпчикову діаграму, на якій відображаються 10 найпопулярніших артистів на Spotify у 2023 році за кількістю слухачів. 
artists = data.groupby("artist(s)_name")["in_spotify_playlists"].sum().nlargest(10)
plt.figure(figsize=(10, 6))
artists.plot(kind="bar", color="brown")
plt.xlabel("Артист")
plt.ylabel("Кількість слухачів")
plt.title("10 найпопулярніших артистів на Spotify у 2023 році")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 5. Порівняння популярності трьох треків: Побудуйте горизонтальну стовпчикову діаграму, яка порівнює кількість прослуховувань трьох конкретних треків у 2023 році. 
tracks = data.sample(n=3)
plt.figure(figsize=(10, 4))
plt.barh(tracks["track_name"], tracks["in_spotify_playlists"], color="grey")
plt.xlabel("Кількість прослуховувань")
plt.ylabel("Назва треку")
plt.title("Порівняння популярності трьох треків на Spotify у 2023 році")
plt.gca().invert_yaxis()
plt.show()
```
