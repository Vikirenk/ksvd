```python
# 1 Створіть графік (scatter plot) для даних ірисів (iris dataset)
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]
print(iris_df.head())

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal width (cm)', y='sepal length (cm)', size='petal length (cm)',
                hue='species', palette='Set2', data=iris_df, sizes=(10, 100), legend='full')
plt.xlabel('Ширина чашелистика (cm)')
plt.ylabel('Довжина чашелистика (cm)')
plt.title('Графік ірисів')

plt.show()
```

```python
# 2 Cтворіть розподільчий графік (scatter plot) для даних про кількість медалей
data = {
    'nation': ['USA', 'Germany', 'China', 'USA', 'Germany', 'China'],
    'medal': ['Gold', 'Gold', 'Gold', 'Silver', 'Silver', 'Silver'],
    'count': [46, 27, 26, 37, 23, 18]
}
medals_df = pd.DataFrame(data)
marker_size = 50
plt.figure(figsize=(6, 2))
sns.scatterplot(x='count', y='nation', hue='medal', style='medal',
                palette='Set1', s=marker_size, data=medals_df, legend='full')
plt.xlabel('Загальна кількість медалей')
plt.ylabel('Нація')
plt.title('Розподільчий графік кількості медалей за націями та типами медалей')
plt.show()
```

```python
# 3 Створіть графік (scatter plot) для даних про рахунки та чайові в ресторані
data = {
    'total_bill': [16.99, 10.34, 21.01, 23.68, 24.59],
    'tip': [1.01, 1.66, 3.5, 3.31, 3.61]
}

tips_df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
sns.regplot(x='total_bill', y='tip', data=tips_df, scatter_kws={"s": 100})
plt.xlabel('Загальна сума рахунку')
plt.ylabel('Сума чайових')
plt.title('Зв\'язок між рахунком та чайовими')
plt.show()
```

```python
# 4 Створіть стовпчасту діаграму (bar chart) для даних про країну Канада зі змінними "рік" (year) та "населення" (population)
years = [2000, 2005, 2010, 2015, 2020]
population = [30412500, 32051800, 34126000, 36541000, 37742154]
plt.figure(figsize=(8, 6))
plt.bar(years, population, color='skyblue')
plt.xlabel('Рік')
plt.ylabel('Населення Канади')
plt.title('Зміна населення Канади за роки')
plt.show()
```

```python
# 5 Створіть кругову діаграму (pie chart) для наданих даних з довільними значеннями "tip" та категоріями "day" (день)
tips = [20, 15, 25, 18]
days = ['Понеділок', 'Вівторок', 'Середа', 'Четвер']
plt.figure(figsize=(8, 8))
plt.pie(tips, labels=days, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'yellowgreen', 'coral', 'gold'])
plt.title('Чайові за дні тижня')
plt.show()
```
