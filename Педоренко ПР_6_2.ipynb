```python
# 1. Побудуйте звичайну лінійну діаграму на основі  pd.DataFrame з двома стовпцями за допомогою бібліотеки Altair для функції y=xcosx для значень x в діапазоні від 0 до 99. 
import pandas as pd
import altair as alt
import numpy as np
x = np.arange(0, 100)
y = x * np.cos(x)
data = pd.DataFrame({'x': x, 'y': y})
chart = alt.Chart(data).mark_line().encode(
    x='x',
    y='y'
).properties(
    title='Графік функції y = x * cos(x)',
    width=600,
    height=400
).configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_title(
    fontSize=16
)
chart = chart.encode(
    x=alt.X('x', axis=alt.Axis(title='Значення x')),
    y=alt.Y('y', axis=alt.Axis(title='Значення y'))
)
chart
```

```python
# 2. Створіть горизонтальну діаграму з мітками на основі даних про врожайність пшениці (data.wheat()) за допомогою бібліотеки  Altair. Задайте вісь x для врожайності пшениці, вісь y для років, а текстові мітки відображайте значення врожайності пшениці на відповідних роках.  
from vega_datasets import data
source = data.wheat()
chart = alt.Chart(source).mark_bar().encode(
    x=alt.X('wheat:Q', axis=alt.Axis(title='Врожайність пшениці')),
    y=alt.Y('year:O', axis=alt.Axis(title='Рік'), sort=None),
    text='wheat:Q',
).properties(
    title='Врожайність пшениці за роками',
    width=500,
    height=300
)
text = chart.mark_text(
    align='left',
    baseline='middle',    
).encode(
    text='wheat:Q'
)
(chart + text).configure_axis(
    labelFontSize=12,
    titleFontSize=14,
).configure_title(
    fontSize=16
)
```

```python
# 3. Побудуйте гістограму із від'ємними значеннями на основі даних про зміни в зайнятості в США (data.us_employment()) за допомогою бібліотеки Altair. 
from vega_datasets import data
df = data.us_employment()
data = pd.DataFrame(df)
chart = alt.Chart(data).mark_bar().encode(
    x='month:T',
    y=alt.Y('nonfarm_change:Q', axis=alt.Axis(title='Зміни в зайнятості')),
    color=alt.condition(
        alt.datum.nonfarm_change < 0,
        alt.value('red'),  
        alt.value('blue')  
    )
).properties(width=500)
chart
```

```python
# 4. Побудуйте гістограму з групуванням на основі даних про врожай ячменю (data.barley()).
from vega_datasets import data
df = data.barley()
data = pd.DataFrame(df)
chart = alt.Chart(data).mark_bar().encode(
    x='year:N',
    y='sum(yield):Q',
    color='year:N',
    column='site:N'
).properties(width=200)
chart
```

```python
# 5. Створіть багатошарову діаграму на основі даних про виробництво електроенергії в штаті Айова (data.iowa_electricity()).
from vega_datasets import data
df = data.iowa_electricity()
data = pd.DataFrame(df)
chart = alt.Chart(data).mark_bar(opacity=0.7).encode(
    x='year:T',
    y='net_generation:Q',
    color='source:N',
    column='source:N'
).properties(width=600)
chart
```

```python
# 6. Побудуйте лінійний графік з маркерами на основі даних про акції (data.stocks()) .
from vega_datasets import data
df = data.stocks()
data = pd.DataFrame(df)
chart = alt.Chart(data).mark_line(point=True, filled=True, color='white', strokeWidth=2).encode(
    x='date:T',
    y='price:Q',
    color='symbol:N'
).properties(width=600)
chart
```

```python
import pandas as pd
data = pd.read_csv("Global_Education.csv",encoding='latin-1')

print("Розмір датасету:", data.shape)

print("\nТипи даних та пропуски:")
print(data.info())

print("\nКількість дублікатів:", data.duplicated().sum())

print("\nНазви стовпців:")
print(data.columns)

print("\nПерші 5 рядків:")
print(data.head())

print("\nОстанні 5 рядків:")
print(data.tail())

print("\nОсновна статистична інформація:")
print(data.describe())
```

```python
# 1. Описовий аналіз:

print("Перші 5 рядків:")
print(data.head())

print("\nОстанні 5 рядків:")
print(data.tail())

mean_values = data.mean()
print("Середнє значення:")
print(mean_values)

median_values = data.median()
print("\nМедіана:")
print(median_values)

min_values = data.min()
print("\nМінімум:")
print(min_values)

max_values = data.max()
print("\nМаксимум:")
print(max_values)
```

```python
# 2. Робота з числовими даними:
# 1
mean_longitude = data['Longitude'].mean()
print("Середнє значення для Longitude:", mean_longitude)
mean_birth_rate = data['Birth_Rate'].mean()
print("Середнє значення для Birth_Rate:", mean_birth_rate)
std_longitude = data['Longitude'].std()
print("Стандартне відхилення для Longitude:", std_longitude)
std_birth_rate = data['Birth_Rate'].std()
print("Стандартне відхилення для Birth_Rate:", std_birth_rate)
# 2
highest_unemployment_country = data.loc[data['Unemployment_Rate'].idxmax(), 'Countries and areas']
highest_unemployment_rate = data['Unemployment_Rate'].max()
lowest_unemployment_country = data.loc[data['Unemployment_Rate'].idxmin(), 'Countries and areas']
lowest_unemployment_rate = data['Unemployment_Rate'].min()
print("Країна з найвищим рівнем безробіття:", highest_unemployment_country)
print("Найвищий рівень безробіття:", highest_unemployment_rate)
print("\nКраїна з найнижчим рівнем безробіття:", lowest_unemployment_country)
print("Найнижчий рівень безробіття:", lowest_unemployment_rate)
```

```python
# 3. Робота з категоріальними даними: 
unique_countries = data['Countries and areas'].unique()
print(unique_countries)
num_countries = data['Countries and areas'].nunique()
print("Кількість країн у датасеті:", num_countries)
```

```python
# 4. Візуалізація даних: 
import matplotlib.pyplot as plt
import seaborn as sns
# 1
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Gross_Primary_Education_Enrollment', y='Gross_Tertiary_Education_Enrollment', data=data, alpha=0.7)
plt.title('Графік розсіювання між Gross Primary та Gross Tertiary Education Enrollment')
plt.xlabel('Gross Primary Education Enrollment')
plt.ylabel('Gross Tertiary Education Enrollment')
plt.show()

 # 2
literacy_rate_male = data['Youth_15_24_Literacy_Rate_Male']
literacy_rate_female = data['Youth_15_24_Literacy_Rate_Female']
countries = data['Countries and areas']
plt.figure(figsize=(12, 8))
plt.barh(countries, literacy_rate_male, color='blue', label='Чоловіки', alpha=0.7)
plt.barh(countries, literacy_rate_female, color='pink', label='Жінки', alpha=0.7)
plt.xlabel('Рейтинг грамотності')
plt.title('Порівняння рейтингів грамотності чоловіків та жінок за країнами')
plt.legend()
plt.show()
```

```python
# 5. Маніпуляції з даними:
#1 
data['Total_Literacy_Rate'] = data['Youth_15_24_Literacy_Rate_Male'] + data['Youth_15_24_Literacy_Rate_Female']
print(data[['Countries and areas', 'Total_Literacy_Rate']])

#2 
highest_literacy_country = data.loc[data['Total_Literacy_Rate'].idxmax(), 'Countries and areas']
highest_literacy_rate = data['Total_Literacy_Rate'].max()
print("Країна з найвищим загальним рейтингом грамотності:", highest_literacy_country)
print("Найвищий загальний рейтинг грамотності:", highest_literacy_rate)
```
