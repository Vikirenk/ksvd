# Завдання 1

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("shopping_trends.csv")
print(data.head())
```

```python
# Перевірте розмір таблиці, наявність пропусків, тип стовпців.
print("Розмір таблиці:\n", data.shape, "\n")
print("Пропуски:\n", data.isnull().sum(), "\n")
print("Типи стовпців:\n", data.dtypes, "\n")
```

```python
# Перейменуйте назви стовпців: Змініть прописні букви на строчні (маленькі) та пробіл на підкреслення.
data.columns = data.columns.str.lower().str.replace(' ', '_')
data.columns
```

```python
# 1 Гістограма віку (age): 
sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))
sns.histplot(data=data, x="age", bins=30, kde=True)
plt.title("Гістограма віку")
plt.xlabel("Вік")
plt.ylabel("Кількість")
plt.show()
```

```python
# 2 Порівняння розподілу покупок за статтею (item_purchased): 
plt.figure(figsize=(10, 5))
sns.countplot(data=data, y="item_purchased", order=data["item_purchased"].value_counts().index, color="purple")
plt.title("Розподілу покупок за статтею")
plt.xlabel("Кількість покупок")
plt.ylabel("Товари")
plt.show()
```

```python
# 3 Залежність між віком та сумою покупок (purchase_amount): 
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="age", y="purchase_amount_(usd)", alpha=0.6, color="red")
plt.title("Залежність між віком та сумою покупок")
plt.xlabel("Вік")
plt.ylabel("Сума покупок")
plt.show()
```

```python
# 4 Графік розподілу оцінок товарів (review_rating): 
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, y="review_rating", color="pink")
plt.title("Розподіл оцінок товарів")
plt.ylabel("Оцінка")
plt.show()
```

```python
# 5 Категоріальна діаграма для статі (gender): 
plt.figure(figsize=(8, 4))
sns.countplot(data=data, x="gender", color= "violet")
plt.title("Розподіл покупок за статтею")
plt.xlabel("Стать")
plt.ylabel("Кількість покупок")
plt.show()
```

```python
# 6 Залежність між рейтингом і сумою покупок (review_rating і purchase_amount): 
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x="review_rating", y="purchase_amount_(usd)", alpha=0.6, color="orange")
plt.title("Залежність між рейтингом товару та сумою покупок")
plt.xlabel("Рейтинг товару")
plt.ylabel("Сума покупок")
plt.show()
```

```python
# 7 Сезонність покупок (season): 
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x="season", color="grey")
plt.title("Сезонність покупок")
plt.xlabel("Сезон")
plt.ylabel("Кількість покупок")
plt.show()
```

```python
# 8 Графік розподілу розмірів одягу (size): 
plt.figure(figsize=(8, 4))
sns.countplot(data=data, x="size", color="green")
plt.title("Розподіл розмірів одягу")
plt.xlabel("Розмір одягу")
plt.ylabel("Кількість покупок")
plt.show()
```

```python
# 9 Кількість покупок відносно типу оплати (payment_method): 
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x="payment_method", color="yellow")
plt.title("Кількість покупок відносно типу оплати")
plt.xlabel("Тип оплати")
plt.ylabel("Кількість покупок")
plt.xticks(rotation=45)  
plt.show()
```

```python
# 10 Залежність між кількістю попередніх покупок (previous_purchases) і сумою поточної покупки (purchase_amount): 
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="previous_purchases", y="purchase_amount_(usd)", alpha=0.6, color="brown")
plt.title("Залежність між покупками")
plt.xlabel("Кількість попередніх покупок")
plt.ylabel("Сума поточної покупки")
plt.show()
```

```python
# Побудуйте графік поверхні (3D surface plot) для візуалізації залежності між трьома числовими змінними з використанням бібліотеки Seaborn. Налаштуйте відображення графіка, включаючи назву графіку, підписи осей і інші необхідні параметри для кращого розуміння візуалізації. Додайте кольорову легенду. 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection="3d")
surface = ax.plot_surface(x, y, z, cmap="viridis")
ax.set_title("3D Surface Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10) 
plt.show()
```
